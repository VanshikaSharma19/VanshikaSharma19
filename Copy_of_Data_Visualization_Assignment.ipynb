```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set up output to display in the notebook
output_notebook()

# Create x and y values
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph to the figure
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6


output_notebook()


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20,25,30,35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.9, source=source,
       legend_field="fruits", line_color='white',
       fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0


show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Bokeh Histogram", tools="", background_fill_color="white")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

p.y_range.start = 0
p.xaxis.axis_label = 'x'
p.yaxis.axis_label = 'Pr(x)'
p.grid.grid_line_color="white"

show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.io import output_notebook
from bokeh.palettes import Viridis256  # Or any other palette you like

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Reshape data for Bokeh
data = {'x': xx.ravel(),
        'y': yy.ravel(),
        'values': data_heatmap.ravel()}

source = ColumnDataSource(data)

# Create color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Bokeh Heatmap", x_range=(0, 1), y_range=(0, 1),
           tools="", toolbar_location=None)

# Add heatmap rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color={'field': 'values', 'transform': color_mapper},
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
